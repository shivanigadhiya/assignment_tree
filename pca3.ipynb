Q2: What Is Eigen-Decomposition and Its Significance in Linear Algebra?
Eigen-Decomposition is the process of factorizing a matrix into its eigenvalues and eigenvectors.

🔹 Significance:

Diagonalizes matrices, simplifying matrix computations.
Used in PCA, spectral clustering, and more.
Reveals matrix properties, such as stability in systems.

Q7: Geometric Interpretation of Eigenvectors and Eigenvalues
Eigenvectors represent special directions where a transformation scales but does not rotate the vector.
Eigenvalues indicate the scaling factor in those directions.
📌 Example:

In PCA, the principal components (PCs) are eigenvectors of the covariance matrix, and eigenvalues represent the variance in each direction.
Q8: Real-World Applications of Eigen-Decomposition
PCA (Principal Component Analysis) – Used in dimensionality reduction.
Google’s PageRank – Eigenvectors help rank web pages.
Quantum Mechanics – Used in solving Schrödinger’s equation.
Vibration Analysis – Eigenvalues determine natural frequencies.
